```python
df = df.drop(columns=["Unnamed: 0"])
df.columns
```

```python
df = df.dropna(subset=["Question", "Context", "Value"])
df["Question"] = df["Question"].astype(str)
df["Context"]  = df["Context"].astype(str)
df["Value"]    = df["Value"].astype(str)
```

```python
df.to_csv("../data/processed/data_ret_clean.csv", index=False)
```

```python
def answer_in_context(row):
    return row["Value"].strip() in row["Context"]

df2 = pd.read_csv("../data/processed/data_ret_clean.csv")
rate = df2.apply(answer_in_context, axis=1).mean()
rate
```
